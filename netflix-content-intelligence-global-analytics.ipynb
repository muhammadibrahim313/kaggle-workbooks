## üìä Data Engineering & Quality Assessment <a id='data-engineering'></a>

```python
# Load the Netflix dataset
# ========================

df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')

print("üì∫ Dataset Loaded Successfully!")
print("=" * 60)
print(f"üìä Total Records: {df.shape[0]:,}")
print(f"üìã Total Features: {df.shape[1]}")
print(f"üíæ Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print("\nüîç First 3 Records:")
print("=" * 60)
df.head(3)
```

```python
# Data Quality Assessment
# =======================

def assess_data_quality(df):
    """Comprehensive data quality assessment"""
    
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=('Missing Values Heatmap', 'Data Type Distribution', 
                       'Missing Values by Column', 'Data Completeness Score'),
        specs=[[{'type': 'heatmap'}, {'type': 'pie'}],
               [{'type': 'bar'}, {'type': 'indicator'}]]
    )
    
    # Missing values heatmap data
    missing_data = df.isnull().sum()
    missing_pct = (missing_data / len(df) * 100).round(2)
    
    # Create visualizations
    # ... (visualization code)
    
    # Summary statistics
    print("\nüìä DATA QUALITY REPORT")
    print("=" * 60)
    for col in df.columns:
        missing = df[col].isnull().sum()
        if missing > 0:
            pct = (missing / len(df) * 100)
            print(f"‚ùå {col:20s}: {missing:,} missing ({pct:.1f}%)")
        else:
            print(f"‚úÖ {col:20s}: Complete")
    
    completeness_score = (1 - df.isnull().sum().sum() / (len(df) * len(df.columns))) * 100
    print(f"\nüìà Overall Data Completeness: {completeness_score:.1f}%")
    
    return missing_data, missing_pct

missing_data, missing_pct = assess_data_quality(df)
```

```python
# Data Cleaning & Feature Engineering
# ===================================

print("üîß Data Cleaning & Feature Engineering")
print("=" * 60)

# Handle missing values strategically
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['director'] = df['director'].fillna('No Director')
df['cast'] = df['cast'].fillna('No Cast')
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.dropna(subset=['date_added', 'rating'])

# Feature Engineering
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['day_added'] = df['date_added'].dt.day
df['weekday_added'] = df['date_added'].dt.day_name()
df['season_added'] = df['date_added'].dt.month.map({
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
})

# Content age
df['content_age'] = 2024 - df['release_year']
df['years_to_netflix'] = df['year_added'] - df['release_year']

# Extract primary country
df['primary_country'] = df['country'].apply(lambda x: x.split(',')[0] if pd.notna(x) else 'Unknown')

# Duration processing
df['duration_value'] = df['duration'].apply(lambda x: int(x.split()[0]) if pd.notna(x) else 0)
df['duration_type'] = df['duration'].apply(lambda x: x.split()[1] if pd.notna(x) and len(x.split()) > 1 else 'Unknown')

# Genre extraction
df['genre_count'] = df['listed_in'].apply(lambda x: len(x.split(',')) if pd.notna(x) else 0)
df['primary_genre'] = df['listed_in'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else 'Unknown')

# Rating categories
rating_groups = {
    'Kids': ['TV-Y', 'TV-Y7', 'TV-G', 'G', 'TV-Y7-FV'],
    'Teens': ['TV-PG', 'PG', 'PG-13', 'TV-14'],
    'Adults': ['TV-MA', 'R', 'NC-17', 'NR', 'UR']
}

def categorize_rating(rating):
    for category, ratings in rating_groups.items():
        if rating in ratings:
            return category
    return 'Unknown'

df['rating_category'] = df['rating'].apply(categorize_rating)

print("‚úÖ Data cleaning completed!")
print(f"‚úÖ {len(df.columns) - 12} new features engineered!")
print(f"‚úÖ Final dataset shape: {df.shape}")
```

```python
# Genre Analysis
# =============

# Extract all genres
all_genres = []
for genres in df['listed_in'].dropna():
    all_genres.extend([g.strip() for g in genres.split(',')])

genre_counts = pd.Series(all_genres).value_counts().head(20)

# Create genre visualization
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# 1. Horizontal bar chart
bars = axes[0].barh(range(len(genre_counts)), genre_counts.values,
                    color=[NETFLIX_RED if i < 3 else NETFLIX_GREY for i in range(len(genre_counts))])
axes[0].set_yticks(range(len(genre_counts)))
axes[0].set_yticklabels(genre_counts.index, fontsize=10)
axes[0].set_xlabel('Count', fontsize=12)
axes[0].set_title('üé¨ Top 20 Netflix Genres', fontsize=14, fontweight='bold', pad=20)
axes[0].grid(axis='x', alpha=0.3, linestyle='--')

# Add value labels
for i, (bar, value) in enumerate(zip(bars, genre_counts.values)):
    axes[0].text(value + 20, bar.get_y() + bar.get_height()/2,
                f'{value:,}', va='center', fontweight='bold' if i < 3 else 'normal')

# 2. Genre Network/Word Cloud placeholder
from wordcloud import WordCloud

genre_text = ' '.join(all_genres)
wordcloud = WordCloud(width=800, height=400, 
                      background_color='white',
                      colormap='Reds',
                      max_words=50).generate(genre_text)

axes[1].imshow(wordcloud, interpolation='bilinear')
axes[1].axis('off')
axes[1].set_title(' Genre Word Cloud', fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

# Genre Correlation Matrix
print("\nüîó Genre Co-occurrence Analysis:")
print("=" * 50)

# Create binary genre matrix
mlb = MultiLabelBinarizer()
genre_matrix = mlb.fit_transform(df['listed_in'].apply(lambda x: x.split(', ') if pd.notna(x) else []))
genre_df = pd.DataFrame(genre_matrix, columns=mlb.classes_)

# Calculate correlation
genre_corr = genre_df.corr()

# Find strongest correlations
strong_corr = []
for i in range(len(genre_corr.columns)):
    for j in range(i+1, len(genre_corr.columns)):
        if abs(genre_corr.iloc[i, j]) > 0.3:
            strong_corr.append((genre_corr.columns[i], genre_corr.columns[j], genre_corr.iloc[i, j]))

strong_corr.sort(key=lambda x: abs(x[2]), reverse=True)

print("Top Genre Correlations:")
for g1, g2, corr in strong_corr[:5]:
    print(f"  ‚Ä¢ {g1} ‚Üî {g2}: {corr:.3f}")
```

## ü§ñ Content Recommendation System <a id='recommendation'></a>

```python
# Advanced Content-Based Recommendation System
# ============================================

print("ü§ñ Building Advanced Recommendation System")
print("=" * 50)

# Prepare features for recommendation
df_rec = df.copy()

# Clean and prepare text features
df_rec['director'] = df_rec['director'].fillna('')
df_rec['cast'] = df_rec['cast'].fillna('')
df_rec['country'] = df_rec['country'].fillna('')
df_rec['listed_in'] = df_rec['listed_in'].fillna('')
df_rec['description'] = df_rec['description'].fillna('')

# Create combined features
df_rec['combined_features'] = (df_rec['director'] + ' ' + 
                               df_rec['cast'] + ' ' + 
                               df_rec['country'] + ' ' + 
                               df_rec['listed_in'] + ' ' + 
                               df_rec['description'] + ' ' +
                               df_rec['rating']).apply(lambda x: x.lower())

# TF-IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(df_rec['combined_features'])

# Calculate similarity matrix
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create recommendation function
def get_recommendations(title, n_recommendations=10):
    """Get content recommendations based on similarity"""
    
    # Find the index of the title
    idx = df_rec[df_rec['title'] == title].index[0]
    
    # Get similarity scores
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get top recommendations (excluding the title itself)
    sim_scores = sim_scores[1:n_recommendations+1]
    
    # Get titles and scores
    movie_indices = [i[0] for i in sim_scores]
    recommendations = df_rec.iloc[movie_indices][['title', 'type', 'listed_in', 'rating']]
    recommendations['similarity_score'] = [i[1] for i in sim_scores]
    
    return recommendations

# Test the recommendation system
print("\nüé¨ Testing Recommendation System")
print("=" * 50)

test_titles = ['Stranger Things', 'The Crown', 'Money Heist']

for test_title in test_titles:
    if test_title in df_rec['title'].values:
        print(f"\nüì∫ Recommendations for '{test_title}':")
        print("-" * 40)
        recommendations = get_recommendations(test_title, 5)
        for idx, row in recommendations.iterrows():
            print(f"  {row['title'][:40]:40s} | Score: {row['similarity_score']:.3f}")

print("\n‚úÖ Recommendation system built successfully!")
```
