    <div class="trend">‚û§ ACF (Auto-Correlation Function) :</div>
    The Auto-Correlation Function (ACF) measures the correlation between a time series and its lagged values. Peaks in the ACF plot indicate significant correlations at those lags, helping identify patterns like seasonality and trends. Positive autocorrelation at lag 1 indicates a positive correlation with the preceding value, while negative autocorrelation suggests an oscillatory pattern.

<div class="trend">‚û§ PACF (Partial Auto-Correlation Function) :</div>
    
   The Partial Auto-Correlation Function (PACF) measures the direct relationship between a time series and its lagged values after removing intermediate effects. Significant spikes in the PACF plot indicate direct relationships between observations at those lags. PACF helps identify autoregressive relationships in the data, aiding in forecasting and modeling.
    
</div>





```python
# Plotting ACF and PACF
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.figure(figsize=(12, 6))
plot_acf(df['Close'], lags=50, ax=plt.subplot(2, 1, 1), color='purple')
plot_pacf(df['Close'], lags=50, ax=plt.subplot(2, 1, 2), color='purple')
plt.show()
```

    <div class="trend">‚û§ The Augmented Dickey-Fuller (ADF) :</div>
    The Augmented Dickey-Fuller (ADF) test is a statistical method used to determine if a time series dataset is stationary or non-stationary. Stationary data has constant statistical properties over time, while non-stationary data does not. The ADF test produces two main outputs: the ADF statistic and the p-value. The ADF statistic is a negative number, and the more negative it is, the stronger the evidence against non-stationarity.

<div class="trend">‚û§ P -Value :</div>
    
   The p-value is like a score that tells us how likely our results are if there's no real effect. A low score means our results are unlikely under that assumption, suggesting there's probably something real going on. A high score means our results are likely even if there's no real effect, so we can't be confident there's anything significant happening. It helps researchers decide if their findings are meaningful or just due to chance.
    
</div>





```python
# Check for stationarity using Augmented Dickey-Fuller test

result = adfuller(df['Close'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
from statsmodels.tsa.arima.model import ARIMA
# Differencing the closing prices
df['Close_diff'] = df['Close'] - df['Close'].shift(1)
df.dropna(inplace=True)

# Performing ADF test on differenced data
result_diff = adfuller(df['Close_diff'])
print('ADF Statistic (After differencing):', result_diff[0])
print('p-value (After differencing):', result_diff[1])
```

<div class="decomposition-box">
    <div class="trend">‚û§ ARIMA  :</div>
    ARIMA, short for Autoregressive Integrated Moving Average, is a popular time series forecasting model that examines and predicts future data points based on past values, trends, and random shocks. It's comprised of three main components: autoregression (AR), differencing (I), and moving average (MA). ARIMA is effective for analyzing stationary or nearly stationary data, as it can adjust for trends and seasonality by differencing the data to make it stationary. This model is widely used in various fields for its ability to capture the underlying patterns and make accurate predictions from historical data.

<div class="trend">‚û§ SARIMAX :</div>
    
   SARIMAX, or Seasonal Autoregressive Integrated Moving Average with Exogenous Factors, is an extension of the ARIMA model that includes additional features to handle seasonal patterns and external factors. In SARIMAX, the 'X' stands for exogenous variables, which are external factors that can influence the time series data but are not part of it. This model is particularly useful when dealing with time series data that exhibits seasonal fluctuations or when there are external factors impacting the data. By incorporating both seasonal components and exogenous variables, SARIMAX provides more flexibility and accuracy in forecasting compared to traditional ARIMA models. It's commonly used in various fields such as economics, finance, and environmental studies to make predictions that account for both internal patterns and external influences.
    
</div>
